```python
import cv2
```

```python
img = cv2.imread('G:\ml.jpeg')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
plt.show()
```

```python
# switching bgr to rgb

new_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(new_img)
plt.show()
```

```python
print(new_img) ## 3-d matrix having three channels of R,G and B. 
```

```python
# now displaying images using open cv library
cv2.imshow('ML',new_img)

cv2.waitKey(25)
cv2.destroyAllWindows()  # BGR TO BGR CONVERSION DUE TO DISPLAYING USING OPEN CV
```
